```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
import pandas as pd
tips = pd.read_csv("Data/tipsdataset.csv")
```

```python
#univariate
tips
```

```python
plt.plot(tips['total_bill'])
plt.show()
```

```python
plt.scatter(tips['size'],tips['total_bill'],s=tips['tip'])
plt.title("Scatter")
plt.show()
```

```python
plt.pie(
    tips['gender'].value_counts(),
    labels=tips['gender'].value_counts().index,
    autopct='%1.1f%%',
    shadow=True, startangle=90
)
plt.show()
```

```python
plt.hist(tips['size'],bins=4) #univar num
plt.show()
```

```python
# Create a visualization
sns.relplot(
data=tips,
x="total_bill", y="tip", col="time",
hue="smoker", style="smoker", size="size",
)
plt.show()
```

```python
sns.relplot(data=tips,x='tip',y='size',col='gender',hue='smoker',size='total_bill')
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True) #uni var (only x)
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", kde=True) #uni var (only x)
sns.histplot(data=tips, x="tip", kde=True) 
```

```python
sns.displot(data=tips, x="gender", kde=True) #uni var (only x)
#sns.displot(data=tips, x="tip", kde=True) 
```
